## Data Cleaning

```python
teleco_data = teleco_base_data.copy()
```

```python
teleco_data.TotalCharges = pd.to_numeric(teleco_data.TotalCharges, errors ='coerce') #to convert total charges to numeric data
teleco_data.isnull().sum()
```

```python
teleco_data.loc[teleco_data['TotalCharges'].isnull()==True]
```

Since the % of these records compared to total dataset is very low , it is safe to ingore them for further processing 

```python
teleco_data.dropna(how='any',inplace=True)
```

```python
print(teleco_data['tenure'].max())
```

**Divinding** customers into bin based on tenure

```python
labels=["{0}-{1}".format(i,i+11) for i in range(1,72,12)]

teleco_data['tenure_group'] = pd.cut(teleco_data.tenure, range(1,80,12), right=False, labels = labels)
```

```python
teleco_data['tenure_group'].value_counts(0)
```

```python
#drop column customerID and tenure
teleco_data.drop(columns= ['customerID','tenure'], axis=1, inplace=True)
teleco_data.head()
```

# Data Exploration

## Univariate Analysis

```python
for i, predictor in enumerate(teleco_data.drop(columns=['Churn', 'TotalCharges', 'MonthlyCharges'])):
    plt.figure(i)
    srs.countplot(data=teleco_data, x=predictor, hue='Churn')
```

```python
teleco_data['Churn'] = np.where(teleco_data.Churn == 'Yes',1,0)
```

```python
teleco_data.head()
```

```python
teleco_data_dummies = pd.get_dummies(teleco_data) #converting all categorical variable into dummies variables
teleco_data_dummies.head()
```

```python
srs.lmplot(data=teleco_data_dummies, x='MonthlyCharges', y='TotalCharges', fit_reg = False) #relationship between monthly charges and total charges
```

churn by monthly charges and total charges


```python
mth = srs.kdeplot(teleco_data_dummies.MonthlyCharges[(teleco_data_dummies["Churn"] ==0)], color="Yellow", shade = True)
mth = srs.kdeplot(teleco_data_dummies.MonthlyCharges[(teleco_data_dummies["Churn"] ==1)], ax=mth, color="Green", shade = True)
mth.legend(["No Churn", "Churn"], loc='upper right')
mth.set_ylabel('Destiny')
mth.set_xlabel('Mothly Charges')
mth.set_title('Mothly Charges by churn')
```

Churn is High when Monthly Charges are high

```python
Tot = srs.kdeplot(teleco_data_dummies.TotalCharges[(teleco_data_dummies["Churn"] == 0) ],
                color="Red", shade = True)
Tot = srs.kdeplot(teleco_data_dummies.TotalCharges[(teleco_data_dummies["Churn"] == 1) ],
                ax =Tot, color="Blue", shade= True)
Tot.legend(["No Churn","Churn"],loc='upper right')
Tot.set_ylabel('Density')
Tot.set_xlabel('Total Charges')
Tot.set_title('Total charges by churn')
```

*Surprising insight * as higher Churn at lower Total Charges

However if we combine the insights of 3 parameters i.e. Tenure, Monthly Charges & Total Charges then the picture is bit clear :- Higher Monthly Charge at lower tenure results into lower Total Charge. Hence, all these 3 factors viz Higher Monthly Charge, Lower tenure and Lower Total Charge are linked to High Churn.

```python
plt.figure(figsize=(20,8))
teleco_data_dummies.corr()['Churn'].sort_values(ascending = False). plot(kind='bar')
```

*Derived Insight: *

**HIGH** Churn seen in case of **Month to month contracts, No online security, No Tech support, First year of subscription and Fibre Optics Internet**

**LOW** Churn is seens in case of **Long term contracts, Subscriptions without internet service** and **The customers engaged for 5+ years**

Factors like Gender, Availability of PhoneService and # of multiple lines have alomost NO impact on Churn

```python
plt.figure(figsize=(12,12))
srs.heatmap(teleco_data_dummies.corr(), cmap="Paired")
```

## Bivariate Analysis

```python
new_df1_target0=teleco_data.loc[teleco_data["Churn"]==0]
new_df1_target1=teleco_data.loc[teleco_data["Churn"]==1]
```

```python
def uniplot(df,col,title,hue =None):
    
    srs.set_style('whitegrid')
    srs.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30
    
    
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
    ax = srs.countplot(data = df, x= col, order=df[col].value_counts().index,hue = hue,palette='bright') 
        
    plt.show()
```

```python
uniplot(new_df1_target1,col='Partner',title='Distribution of Gender for Churned Customers',hue='gender')
```

```python
uniplot(new_df1_target0,col='Partner',title='Distribution of Gender for Non Churned Customers',hue='gender')
```

```python
uniplot(new_df1_target1,col='PaymentMethod',title='Distribution of PaymentMethod for Churned Customers',hue='gender')
```

```python
uniplot(new_df1_target1,col='Contract',title='Distribution of Contract for Churned Customers',hue='gender')
```

```python
uniplot(new_df1_target1,col='TechSupport',title='Distribution of TechSupport for Churned Customers',hue='gender')
```

```python
uniplot(new_df1_target1,col='SeniorCitizen',title='Distribution of SeniorCitizen for Churned Customers',hue='gender')
```
